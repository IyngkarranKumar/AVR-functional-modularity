# Exploratory Data and Network Analysis

The aims of this notebook are as follows:

## Data

```python
import numpy as np
import os
import zipfile
import matplotlib.pyplot as plt

import torch
from torch.utils.data import DataLoader,Dataset
import torchvision
```

```python
path = 'RAVEN-10000-release.zip'
if not os.path.isdir('RAVEN-10000'):
    with zipfile.ZipFile(path,mode='r') as dataset:
        dataset.extractall('RAVEN-10000') #extract into dataset
```

```python
from src.utils import dataset #using Zhang's utils.py

path = 'RAVEN-10000/RAVEN-10000/' 
img_size = 160
batch_size = 32

train_set = dataset(dataset_path=path,dataset_type='train',
                    img_size=img_size,test=False)
train_loader = iter(DataLoader(train_set,batch_size = batch_size,
                          shuffle=True))
batch = next(train_loader)
```

```python
#plot first matrix in first batch
matrix = batch[0][1]
n_imgs,_,_=matrix.size()

blank_img = torch.zeros_like(matrix[0]).unsqueeze(0)
Q_matrix = torch.cat([matrix[:8],blank_img]) #8 imgs and one empty
A_matrix = matrix[8:]

fig_Q,axs_Q = plt.subplots(3,3,figsize=(6,6)); fig_Q.suptitle('Question Matrix')
for i,ax in enumerate(axs_Q.ravel()):
    ax.imshow(Q_matrix[i],cmap='gray')
    ax.set_xticks([]); ax.set_yticks([])
    
fig_A,axs_A = plt.subplots(2,4,figsize=(6,6)); fig_A.suptitle('Answers')
for i,ax in enumerate(axs_A.ravel()):
    ax.imshow(A_matrix[i],cmap='gray')
    ax.set_xticks([]); ax.set_yticks([])
fig_A.tight_layout()
```

```python
from src import networks_2
from torchsummary import summary

imgs,targets = next(train_loader)
model = networks_2.CoPINet()

summary(model,imgs.size())
```

```python
from torch.utils import tensorboard

writer = tensorboard.SummaryWriter()

writer.add_graph(model,imgs)
writer.close()
#https://tensorboard.dev/experiment/mGLbkcK2T1OT1jLe5qrOHg/#graphs&run=Oct15_11-30-46_Iyngkarrans-MacBook-Air.local
```
